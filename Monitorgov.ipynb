```python
import re
import sys
import requests
URL = 'https://app.monitorgov.com.br/login'


session = requests.Session()

sign_in_page = session.get(URL).content

for linha in str(sign_in_page).split('\n'):
    
    m = re.search('name="_token" value="([^"]+)"', linha)
    if m:
        break

print(m)
token = None
if m:
    token = m.group(1)

print(token)
if not token:
    print('Sem token')
    sys.exit(1)

data = {'email': 'xx',
        'password': 'xx',
        '_token': token}
monitor = session.post(URL, data=data)
if monitor.status_code != 200:
    print('Error Login')
    sys.exit(1)

#print(r.content)

print(monitor.status_code)
print(monitor.url)
print(monitor.headers)
print(len(monitor.content))
```
